```python
import difflib 

def fuzzy_merge(df1, df2, left_on, right_on, how='inner', cutoff=0.50):
    df_other= df2.copy()
    df_other[left_on] = [get_closest_match(x, df1[left_on], cutoff) 
                         for x in df_other[right_on]]
    return df1.merge(df_other, on=left_on, how=how)

def get_closest_match(x, other, cutoff):
    matches = difflib.get_close_matches(x, other, cutoff=cutoff)
    return matches[0] if matches else x
```

```python
m_t = fuzzy_merge(df1, df2, left_on='product', right_on='product')
```

```python
df = m_t_s_a_w
```

```python
df.head()
```

```python
### For Morrisons
df['unit_price_morrisons_new'] = df['unit_price_morrisons'].apply(lambda x: x.replace('£',""))
df['morrisons'] = df['unit_price_morrisons_new'].apply(lambda x: float(x[0:x.find('p')]))
```

```python
### For Sainsbury
df['sainsbury'] = df['unit_price_sainsbury'].apply(lambda x: x.replace('£',"").replace('p',""))### For Sainsbury
df['sainsbury'] = df['unit_price_sainsbury'].apply(lambda x: x.replace('£',"").replace('p',""))
```

```python
### For Asda
df['unit_price_asda_new'] = df['unit_price_asda'].apply(lambda x: str(x).replace('p',"").replace('£',"")).fillna(0)
df['asda'] = df['unit_price_asda_new'].apply(lambda x: x[1:x.find('/')])
```

```python
### For waitrose
df['unit_price_waitrose_new'] = df['unit_price_waitrose'].apply(lambda x: str(x).replace('p',"").replace('£',"")).fillna(0)
df['waitrose'] = df['unit_price_waitrose_new'].apply(lambda x: x[1:x.find('/')])
```

```python
df.drop([x for x in df.columns if x.endswith('_new')],inplace=True,axis=1)
```
